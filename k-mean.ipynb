```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Sample customer purchase history data (features could include purchase frequency, total amount spent, etc.)
purchase_history = np.array([
    [3, 250],   # Customer 1
    [1, 150],   # Customer 2
    [5, 450],   # Customer 3
    [2, 200],   # Customer 4
    [4, 350],   # Customer 5
    [6, 500]    # Customer 6
])

# Standardize the features
scaler = StandardScaler()
purchase_history_scaled = scaler.fit_transform(purchase_history)

# Specify the number of clusters (K)
num_clusters = 3

# Create and fit the K-means model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(purchase_history_scaled)

# Get cluster centers and labels
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_

# Visualize the clusters
plt.scatter(purchase_history[:, 0], purchase_history[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', c='red', label='Centroids')
plt.xlabel('Purchase Frequency')
plt.ylabel('Total Amount Spent')
plt.title('Customer Segmentation based on Purchase History')
plt.legend()
plt.show()

# Output cluster labels
for i, label in enumerate(cluster_labels):
    print(f"Customer {i+1} belongs to cluster {label+1}")
```
